## Merge

```python
test_data_1 = pd.DataFrame({"Sinif":["A","B","C","D"],
              "Not" : ["40","80","70","90"]
             })
test_data_1
```

```python
test_data_2 = pd.DataFrame({"Sinif":["A","B","C","E"],
              "ogrenci_adi" : ["Can","Buse","Pelin","Semih"]
             })
test_data_2
```
