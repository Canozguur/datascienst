```python
import pandas as pd

# CSV dosyalarını oku
sinif_5A = pd.read_csv("sinif_csvleri/5A_sinifi.csv")
sinif_5B = pd.read_csv("sinif_csvleri/5B_sinifi.csv")

# DataFrame'leri birleştir
tum_ogrenciler = pd.concat([sinif_5A, sinif_5B], ignore_index=True)

# Ortalama kolonunu oluştur
tum_ogrenciler["ortalama"] = (
    tum_ogrenciler["matematik"] +
    tum_ogrenciler["turkce"] +
    tum_ogrenciler["ingilizce"] +
    tum_ogrenciler["beden_egitimi"]
) / 4

# En çalışkan öğrenciyi bul
en_caliskan = tum_ogrenciler.loc[
    tum_ogrenciler["ortalama"].idxmin()
]

print(en_caliskan)
```

```python
import pandas as pd

# CSV dosyalarını oku
sinif_5A = pd.read_csv("sinif_csvleri/5A_sinifi.csv")
sinif_5B = pd.read_csv("sinif_csvleri/5B_sinifi.csv")

# Hangi sınıf olduğunu ekle
sinif_5A["sinif"] = "5A"
sinif_5B["sinif"] = "5B"

# Birleştir
tum_ogrenciler = pd.concat([sinif_5A, sinif_5B], ignore_index=True)

# Ortalama hesapla
tum_ogrenciler["ortalama"] = (
    tum_ogrenciler["matematik"] +
    tum_ogrenciler["turkce"] +
    tum_ogrenciler["ingilizce"] +
    tum_ogrenciler["beden_egitimi"]
) / 4

# En çalışkan öğrenci
en_caliskan = tum_ogrenciler.loc[
    tum_ogrenciler["ortalama"].idxmax()
]
print(en_caliskan)
```
