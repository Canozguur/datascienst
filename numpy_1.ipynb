```python
data = np.random.randint(0,20,(2,3))
df =pd.DataFrame(data,columns=["sayi1","sayi2","sayi3"])
```

```python
df
```

```python
df["toplami"] = (df.sayi1 + df.sayi2 + df.sayi3)
df["ortalama"] = (df.sayi1 + df.sayi2 + df.sayi3)/3
```

```python
df["sayi1"].max()
```

```python
def find_max():
    max = 0
    for i in df.columns:
        print(i)
        for j in df[i]:
            print(j)
            if j > max:
                max = j
    return max
```
