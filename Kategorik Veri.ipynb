```python
data = np.random.randn(1000)
```

```python
ara = pd.qcut(data,4)
ara
```

```python
type(ara)
```

```python
ara = pd.qcut(data,4,labels = ["Q!","Q2","Q3","Q4"])
ara
```

```python
ara = pd.Series(ara,name="ceyrek")
```

```python
pd.Series(data).groupby(ara, observed = False).agg(["count","min","max"]).reset_index()
```
