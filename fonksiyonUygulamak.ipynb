```python
import pandas as pd
import numpy as np
```

```python
vs = pd.DataFrame({"anahtar": list("ABC")*2,
                  "veri1":range(6),
                  "veri2":np.arange(5,11)})
```

```python
vs
```

```python
grup = vs.groupby("anahtar")
```

```python
grup.aggregate(["min","median","max"])
```

```python
grup.agg({"veri1":"min","veri2":"max"})
```

```python
def f(x):
    return x.min() - x.max()
grup.agg(f)
```
