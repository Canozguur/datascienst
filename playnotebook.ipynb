```python
class Find():
    def is_Prime(x):
        if x < 2 : 
            return False
        for i in range(2,x):
            if x % i == 0:
                return False
        return True
    def find_primes_in_array(x):
        array = []
        for i in range(x):
        
            if Find.is_Prime(i):
                array.append(i)
        return array
    def is_perfect_number(x):
        # 6 = 1+ 2+ 3
        total = 0
        for i in range(1,x):
            if x % i == 0:
                total += i
        if x == total :
            return True
        else :
            return False
            
    def is_palindrome(word):

        if word.lower() == word.lower()[::-1]:
            return True

        return False
```

```python
class Student:
    def __init__(self,name,surname):
        self.name = name
        self.surname = surname
    def show_info(self):
        print(f"name :{self.name}")
        print(f"surname :{self.surname}")
```
